```python
def gen_node_file():

#     file = open(os.path.join(pkubase_path, 'pkubase-complete.txt'),'r',encoding='utf-8').readlines()
#     print('数据库原始文件读取完成')
#     csvf = open('node.csv','w',newline='',encoding = 'utf-8')

    
    entity_dic = dict()
#     for i in tqdm(range(len(file))):
    with open(os.path.join(pkubase_path, 'pkubase-complete.txt'),'r', encoding='utf-8') as fin, open(
        os.path.join(pkubase_path, 'node.csv'),'w',newline='',encoding = 'utf-8') as csvf:
        w = csv.writer(csvf)
        w.writerow(("id:ID","name",":LABEL"))
        idx = 0
#         print(len(file))
        for line in fin:
#         if '<类型>' in file[i]:
#             continue
#2019.8.30这次导入将实体的类型属性也导入数据库中
        
            items = line.split('\t')
            if len(items) != 3:
                print(line)
                continue

            entity1 = items[0].split('\t')[0]
            if entity1 not in entity_dic:
                idx+=1
                entity_dic[entity1]=idx
                w.writerow((str(idx),entity1,'Entity'))

            entity2 = items[2].rstrip(' .\n')
            if entity2 not in entity_dic:
                idx+=1
                entity_dic[entity2] = idx
                w.writerow((str(idx),entity2,'Entity'))

    print(len(entity_dic))
#     csvf.close()
    return entity_dic
```

```python
file = open(os.path.join(pkubase_path, 'pkubase-complete.txt'),'r',encoding='utf-8').readlines()
```

```python
atti_dic={}
for i in range(1,31):
    atti_dic[i]={}
```

```python
for i in tqdm(range(len(file))):
    items = file[i].split('\t')
    if len(items) != 3:
        print(file[i])
        continue

    entity1 = items[0]
    entity2 = items[2].rstrip(' .\n')
    try:
        if entity1[0]=='\"':
            entity1=entity1.strip('\"')
            atti_dic[len(entity1)][entity1]=''
        if entity2[0]=='\"':
            entity2=entity2.strip('\"')
            atti_dic[len(entity2)][entity2]=''
    except:
        continue
atti_dic[30]
```

```python
pickle.dump(atti_dic,open(os.path.join(pkubase_path, 'prop_dic.pkl'),'wb'))
```

obtain type dictionary

```python
type_dic={}
for i in tqdm(range(len(file))):
    line = file[i].split('\t')
    if len(line)==3:
        if line[1]=='<类型>':
            entity = line[0]
            ent_type = line[2].rstrip(' .\n')
            if entity in type_dic:
                type_dic[entity].append(ent_type)
            else:
                type_dic[entity]=[ent_type]
```

```python
pickle.dump(type_dic,open(os.path.join(pkubase_path, 'type_dic.pkl'),'wb'))
```
